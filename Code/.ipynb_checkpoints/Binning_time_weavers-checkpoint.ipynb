```python
import os
import sys
sys.path.append(os.path.abspath('..'))

import numpy as np
import pandas as pd
from inference.walker_inference import BiasedPersistentInferer, prepare_paths
from in_silico.sources import PointSource


# Variables needed for Ensemble Monte Carlo
n_iter = 1000
n_walkers = 75


"""
Space and time slices taken from the Weavers paper
space_slice splits the dataframe into different spatial bins. time_slice splits the data into different temporal bins,
this allows for the inference pipeline to calculate the spatial-temporal values for the bias, persistence, and weights. 

def space_slice(theta):
    s35 = theta[(theta['r'] >= 0)  & (theta['r'] <= 70)]
    s105 = theta[(theta['r'] >= 70)  & (theta['r'] <= 140)]
    s195 = theta[(theta['r'] >= 140)  & (theta['r'] <= 250)]
    s305 = theta[(theta['r'] >= 250)  & (theta['r'] <= 360)]
    s430 = theta[(theta['r'] >= 360)  & (theta['r'] <= 500)]
    return [s35,s105,s195,s305,s430]


def time_slice(space):
    t10 = space[(space['t'] >= 0) & (space['t'] <= 1500)]
    t22 = space[(space['t'] >= 1500) & (space['t'] <= 2100)]
    t43 = space[(space['t'] >= 2100) & (space['t'] <= 3500)]
    t58 = space[(space['t'] >= 3500) & (space['t'] <= 3900)]
    t85 = space[(space['t'] >= 3900) & (space['t'] <= 6400)]
    t115 = space[(space['t'] >= 6400) & (space['t'] <= 7500)]
    t143 = space[(space['t'] >= 7500) & (space['t'] <= 9600)]
    t178 = space[(space['t'] >= 9600) & (space['t'] <= 11700)]
    times = [t10,t22,t43,t58,t85,t115,t143,t178]
    return times

"""


def space_slice(theta):
    r_bins = [0, 70, 140, 250, 360, 560]
    space_slices = np.digitize(theta['r'], r_bins)
    return [theta[space_slices == i] for i in range(1, len(r_bins))]

def time_slice(space):
    t_bins = [0, 1500, 2100, 3500, 3900, 6400, 7500, 9600, 11700]
    time_slices = np.digitize(space['t'], t_bins)
    return [space[time_slices == i] for i in range(1, len(t_bins))]

"""
First, we slice the data into their respective spatial bins and then we split the spatial bins further to bin them in 
their respective temporal bins. When then run the inference pipeline on the space_time_dataframe. 
"""
trajectory = np.load('../data/Trajectory_dataframes/new_2023_01_27-07:32:27_AM_control',allow_pickle=True)
distance = space_slice(trajectory)
times = ["t10","t22","t43","t58","t78","t108","t143","t178"]
space_time_dataframe = [time_slice(distance[i]) for i in range(len(distance))]
```

```python
space_time_dataframe[4][-1]
```

```python

"""
This will run the inference method iteratively for each temporal and spatial bin and save
them as a numpy array for analysis in the data analysis Python script.
"""
# Error term to make sure we catch any issues where PointSource isn't used as the source


source = PointSource(position=np.array([0, 0]))
"""
if source != PointSource:
    print("Cannot run bias persistence inference pipeline, please check that source is PointSource")
else:
"""
k = 0
time = space_time_dataframe[0]

for i in range(len(space_time_dataframe)):
    for j in range(len(time)):
        k += 1  # Tracks the number of bins
        print('analysing bin {}/{}'.format(k, (len(distance) * len(time))))  # to give an overall sense of progress
        inferer = BiasedPersistentInferer(
            prepare_paths(
                [paths[['x', 'y']].values for id, paths in space_time_dataframe[i][j].groupby('trackID')],
                include_t=False), source)
        inf_out = inferer.ensembleinfer(n_walkers, n_iter)
        np.save(f'../data/WalkerData/PosterData/New_Control_wound_{i}{j}', inf_out)
```

```python
data = inf_out[0].get_chain(flat=True)
```

```python
W = data[500:,0]
P = data[500:,1]
B = data[500:,2]
```

```python
import matplotlib.pyplot as plt
plt.hist(W,label='$W$ = {:.2f} $\pm$ {:.2f}'.format(np.mean(W), np.std(W)),bins=100,alpha=0.6,density=True)
plt.hist(P,label='$W$ = {:.2f} $\pm$ {:.2f}'.format(np.mean(W), np.std(W)),bins=100,alpha=0.6,density=True)
plt.hist(B,label='$W$ = {:.2f} $\pm$ {:.2f}'.format(np.mean(W), np.std(W)),bins=100,alpha=0.6,density=True)

plt.show()
```
